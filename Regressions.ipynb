```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/world-happiness-report/2022.csv") #import Dataset
df = pd.DataFrame(data)
```

```python
df2 = df.dropna()
```

```python
# the following columns are string and numbers are devided by a comma
# so we must remove comma
# then convert them to int

df2['Explained by: GDP per capita']=df2['Explained by: GDP per capita'].str.replace(',','')
df2['Explained by: Healthy life expectancy']=df2['Explained by: Healthy life expectancy'].str.replace(',','')
df2['Explained by: Social support']=df2['Explained by: Social support'].str.replace(',','')
df2['Explained by: Freedom to make life choices']=df2['Explained by: Freedom to make life choices'].str.replace(',','')
df2['Explained by: Generosity']=df2['Explained by: Generosity'].str.replace(',','')
df2['Explained by: Perceptions of corruption']=df2['Explained by: Perceptions of corruption'].str.replace(',','')
df2['Dystopia (1.83) + residual']=df2['Dystopia (1.83) + residual'].str.replace(',','')

df2["Explained by: GDP per capita"] = df2['Explained by: GDP per capita'].astype(int)
df2['Explained by: Healthy life expectancy']=df2['Explained by: Healthy life expectancy'].astype(int)
df2['Explained by: Social support']=df2['Explained by: Social support'].astype(int)
df2['Explained by: Freedom to make life choices']=df2['Explained by: Freedom to make life choices'].astype(int)
df2['Explained by: Generosity']=df2['Explained by: Generosity'].astype(int)
df2['Explained by: Perceptions of corruption']=df2['Explained by: Perceptions of corruption'].astype(int)
df2['Dystopia (1.83) + residual']=df2['Dystopia (1.83) + residual'].astype(int)
df2
```

```python
# 3 rows have zero values

df2 = df2[df2["Explained by: Social support"]>0]
df2 = df2[df2["Explained by: GDP per capita"]>0]
df2 = df2[df2["Explained by: Healthy life expectancy"]>0]
```
